```python
import cv2
import os
import numpy as np
from PIL import Image
```

```python
image_path = r"C:\Users\1603585\Documents\Python Scripts\Aishwarya\GAN_pix2pix\pytorch-CycleGAN-and-pix2pix-master\datasets\docdata_resized\train\\"
output_path = r"C:\Users\1603585\Documents\Python Scripts\Aishwarya\GAN_pix2pix\pytorch-CycleGAN-and-pix2pix-master\datasets\docdata_resized\image_concate\\"
```

```python
w_r = 256
w_c = 256
```

```python
for file in os.listdir(image_path):
    image = cv2.imread(image_path + file)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print(gray.shape)
    count = 0
    row_append = []
    image_dict = {}
    
    for r in range(0,gray.shape[0], w_r):
        for c in range(0,gray.shape[1], w_c):
            window = gray[r:r+w_r,c:c+w_c]
            row_append.append(window)
        count+=1
        image_dict[count] = row_append
        row_append = []
        
#     print(count)
    row_img = []
    for k,v in image_dict.items():
#         print(k)
        col_img = np.concatenate([v[0], v[1],v[2],v[3]], 1)
        row_img.append(col_img)
    final_img =  np.concatenate([row_img[0], row_img[1],row_img[2],row_img[3]], 0)  
#     final_img.shape
    cv2.imwrite(output_path + file,final_img)
    
#     break
```
